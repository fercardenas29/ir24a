```python
# all terms
# my diccionary
terms = set()
for document in docs:
    terms.update(document.lower().split())
```

```python
matrix_frecuency = []
for document in docs:
    # count frecuency
    count = Counter(document.lower().split())
    #create frecuency
    frequency = [count.get(word, 0) for word in terms]
    #frecuency in matrix
    matrix_frecuency.append(frequency)
```

```python
#matrix query
matrix_query = []

for document in query:
    # count frecuency
    count = Counter(document.lower().split())
    #create frecuency
    frequency = [count.get(word, 0) for word in terms]
    #frecuency in matrix
    matrix_query.append(frequency)
```

```python
for fila in matrix_query:
    print(fila)
```

```python
distance = cosine_similarity(matrix_frecuency, matrix_query)
```

```python
print(distance)
```

```python
len(distance)
```

```python
import numpy as np

top_3 = []
for fila in distance:
  index = np.argsort(fila)[-3:][::-1]
  top_3.append(fila[index])
```

```python
print(top_3)
```

```python
len(top_3)
```

```python
query_matrix = np.array(query)
top_3_matrix = np.array(top_3)
```

```python
concatenated_matrix = np.empty((len(query_matrix), 2), dtype=object)
```

```python
for i in range(len(query_matrix)):
    concatenated_matrix[i] = [query_matrix[i], top_3_matrix[i]]
```

```python
print(concatenated_matrix)
```

```python
# create matrix 
name_file = "matrix_query_and_top3.csv"

# write in csv
with open(name_file, "w", newline="", encoding="utf-8") as archivo_csv:
    writer_csv = csv.writer(archivo_csv)
    for fila in concatenated_matrix:
        writer_csv.writerow(fila)
```
